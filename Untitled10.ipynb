```python
import numpy as np

# Leaky Integrate-and-Fire neyron modeli
class LIFNeuron:
    def __init__(self, threshold, reset_value, decay_factor, refractory_period):
        self.threshold = threshold
        self.membrane_potential = 0
        self.decay_factor = decay_factor # Potensialning pasayish koeffitsienti
        self.refractory_period = refractory_period # Refrakter davr
        self.refractory_end_time = -1
        self.spike_time = -1 # Initialize spike_time

    def update(self, incoming_spikes, current_time):
        # Refrakter davrda neyron ishlamaydi
        if current_time < self.refractory_end_time:
            self.spike_time = -1 # Reset spike_time if in refractory period
            return False

        # Potensialni yangilash (yig'ish va kamayish)
        self.membrane_potential = (self.membrane_potential * self.decay_factor) + np.sum(incoming_spikes)

        # Spayk (impuls) hosil bo'lishi
        if self.membrane_potential >= self.threshold:
            self.membrane_potential = 0 # Reset
            self.refractory_end_time = current_time + self.refractory_period
            self.spike_time = current_time # Record spike time
            return True
        self.spike_time = -1 # No spike
        return False

# Sinaps sinfi (og'irlik)
class Synapse:
    def __init__(self, weight):
        self.weight = weight
# Spike-Timing-Dependent Plasticity (STDP)
def stdp(pre_spike_time, post_spike_time, weight, learning_rate, tau_positive, tau_negative):
    if pre_spike_time > 0 and post_spike_time > 0: # Only apply STDP if both pre- and post-synaptic neurons spiked
        delta_t = post_spike_time - pre_spike_time
        if delta_t > 0:
            return weight + learning_rate * np.exp(-delta_t / tau_positive)
        else:
            return weight - learning_rate * np.exp(delta_t / tau_negative)
    return weight
# Simulation Parameters
time_steps = 100
input_size = 5
hidden_size = 3
output_size = 1

# Network Initialization
input_neurons = [LIFNeuron(threshold=1.0, reset_value=0.0, decay_factor=0.9, refractory_period=2) for _ in range(input_size)]
hidden_neurons = [LIFNeuron(threshold=1.0, reset_value=0.0, decay_factor=0.9, refractory_period=2) for _ in range(hidden_size)]
output_neurons = [LIFNeuron(threshold=1.0, reset_value=0.0, decay_factor=0.9, refractory_period=2) for _ in range(output_size)]

input_to_hidden_synapses = np.random.rand(input_size, hidden_size)
hidden_to_output_synapses = np.random.rand(hidden_size, output_size)

learning_rate = 0.01
tau_positive = 20
tau_negative = 20
# Spike Train Pattern to Detect
pattern = [1, 0, 1, 0, 1]
# Simulation Loop
for t in range(time_steps):
    # Kirish signallarini generatsiya qilish
    input_spikes = np.random.randint(0, 2, size=input_size)

    # Neyronlarni yangilash (Kirish -> Yashirin -> Chiqish)
    hidden_spikes_activations = np.zeros(hidden_size)
    for i, neuron in enumerate(input_neurons):
        # For input neurons, the 'incoming_spikes' is just the input_spikes[i]
        # The update method now also records the spike_time if it fires
        fired = neuron.update(input_spikes[i], t) # Pass only the input spike to the input neuron
        if fired:
            # If an input neuron fires, its spike contributes to all connected hidden neurons
            hidden_spikes_activations += input_to_hidden_synapses[i, :]

    output_spikes_activations = np.zeros(output_size)
    for j, neuron in enumerate(hidden_neurons):
        # Update hidden neurons with their accumulated activations
        fired = neuron.update(hidden_spikes_activations[j], t)
        if fired:
            # If a hidden neuron fires, its spike contributes to all connected output neurons
            output_spikes_activations += hidden_to_output_synapses[j, :]

    for k, neuron in enumerate(output_neurons):
        neuron.update(output_spikes_activations[k], t)

    # STDP Learning (Sinapslarni yangilash)
    for i in range(input_size):
        for j in range(hidden_size):
            input_to_hidden_synapses[i, j] = stdp(input_neurons[i].spike_time, hidden_neurons[j].spike_time,
                                                input_to_hidden_synapses[i, j], learning_rate, tau_positive, tau_negative)

    for j in range(hidden_size):
        for k in range(output_size):
            hidden_to_output_synapses[j, k] = stdp(hidden_neurons[j].spike_time, output_neurons[k].spike_time,
                                                 hidden_to_output_synapses[j, k], learning_rate, tau_positive, tau_negative)

    # Na'munani aniqlash (Pattern detection)
    # Check if the pattern matches based on recent spikes from input neurons
    # Only consider neurons that are part of the pattern (pat == 1)
    pattern_detected = True
    for neuron, pat in zip(input_neurons, pattern):
        if pat == 1 and neuron.spike_time != t:
            pattern_detected = False
            break
        elif pat == 0 and neuron.spike_time == t:
            pattern_detected = False
            break

    if pattern_detected:
        print(f"Pattern detected at time step {t}")
```
